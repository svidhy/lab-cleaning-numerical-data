```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### Instructions
1. Use the numerical variables from our dataset this week.
2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
numerical = pd.read_csv('numerical.csv')
```

```python
numerical.head()
```

```python
#Dropping the unnamed column(default index column from csv)
numerical = numerical.drop(['Unnamed: 0'], axis=1)
```

```python
numerical.head()
```

```python
numerical.shape
```

```python
for column in numerical.columns:
    sns.histplot(x=numerical[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.show()
    
    sns.displot(x=np.log(numerical[column]), kde=True)
    plt.title(f'Log Transformed Histogram of {column}')
    plt.show()
```

```python
for column in numerical.columns:
    sns.histplot(x=numerical[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.show()
    
    sns.displot(x=np.sqrt(numerical[column]), kde=True)
    plt.title(f'Square Root Transformed Histogram of {column}')
    plt.show()
```

```python
for column in numerical.columns:
    sns.histplot(x=numerical[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.show()
    
    sns.displot(x=(numerical[column]**(1/1.2)), kde=True)
    plt.title(f'Exponentially Transformed Histogram of {column}')
    plt.show()
```

### Conclusion

As can be seen most of the columns cannot be classified as normal distribution. The transformations helped to an extent but they didn't form the true normal distribution. 



| Ind  | Column  | Histogram  | Log Transformed  | Square Root Transformed  |
|---|---|---|---|---|
| 1  | customer_lifetime_value  | Exponential distribution | Transformed but not normal  | Similar to Log  |
| 2  | income  |  Not exponential | Closer to exponential  | Neither normal or exponential  |
| 3  | monthly_premium_auto  |  Closer to exponential | Not much difference in shape  |   Similar to log|
| 4  | months_since_last_claim  |   |  Closer to exponential | Similar to log  |
| 5  | months_since_policy_inception  | Close to normal but not truly  |  Exponential | Exponential  |
| 6  | number_of_open_complaints   | Close to exponential  | Close to exponential  | Close to exponential  |
